```python
from sklearn import tree
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
dataframe = pd.read_csv("../input/airline-passenger-satisfaction/train.csv")
```

```python
dataframe = pd.get_dummies(dataframe) # Converts string inputs into numerical values
```

```python
X = dataframe.drop(["Gender_Female", "Gender_Male", "Age", "id", "satisfaction_satisfied", "satisfaction_neutral or dissatisfied"], axis=1).to_numpy()
X = X.astype(int)
Y1 = dataframe[:][["satisfaction_satisfied", "satisfaction_neutral or dissatisfied"]].to_numpy()
```

```python
Y = np.zeros(shape=(103904,))
for i in range(103904):
    if Y1[i][0] == 1: # Satisfactory (1)
        Y[i] = 0
    else:
        Y[i] = 1
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size=10)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
classifier = tree.DecisionTreeClassifier()
classifier = classifier.fit(xtrain, ytrain)
```

```python
predicted = classifier.predict(xtest)
```

```python
acc = accuracy_score(ytest, predicted)
print("Accuracy %.5f" % acc)
```
